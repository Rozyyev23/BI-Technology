```python
import sklearn
```

```python
from sklearn import preprocessing 
label = preprocessing.LabelEncoder()
```

```python
col = ['address', 'info', 'z_address', 'bathrooms', 'bedrooms',
       'finishedsqft', 'lastsolddate', 'lastsoldprice', 'latitude',
       'longitude', 'neighborhood', 'totalrooms', 'usecode', 'yearbuilt',
       'zestimate', 'zindexvalue', 'zipcode', 'zpid']
```

```python
for i in col:
  df2[i]= label.fit_transform(df2[i]) 
  print(df2[i].unique())
```

```python
y = df2['yearbuilt']
```

```python
X = df2[['address', 'bathrooms', 'bedrooms',
       'finishedsqft', 'lastsolddate', 'lastsoldprice',
        'totalrooms', 'usecode']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle = True)
```

```python
from sklearn.linear_model import LogisticRegression #импорт классификатора
log = LogisticRegression()
log.fit(X_train,y_train) #обучение модели
print(log.score(X_train, y_train))
print(log.score(X_test, y_test))
```

```python
y_pred = log.predict(X_test)
print(sklearn.metrics.classification_report(y_test,y_pred))
```

```python
from sklearn import metrics 
```

```python
# Метод k-ближайших соседей
import sklearn.neighbors as neigh

kn = neigh.KNeighborsClassifier(n_neighbors=19).fit(X_train,y_train)
y_pred = kn.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
# Метод опорных векторов SVM
import sklearn.svm as svm

svect = svm.LinearSVC(max_iter=19).fit(X_train, y_train)
y_pred = svect.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
# Метод опорных векторов SVM
import sklearn.svm as svm

svect = svm.LinearSVC(max_iter=91).fit(X_train, y_train)
y_pred = svect.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```
